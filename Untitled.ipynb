```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": False,
    "mathtext.fontset": "dejavuserif"})
# ## for Palatino and other serif fonts use:
# plt.rcParams.update({
#     "text.usetex": True,
#     "font.family": "serif",
#     "font.serif": ["Palatino"],
# })
```

```python
def BAF10(wl):
    #BAF10
    n2 = 1 + (1.5851495*wl**2)/(wl**2-0.00926681282)+ \
         (0.143559385*wl**2)/(wl**2 - 0.0424489805) + \
         (1.08521269*wl**2)/(wl**2-105.613573)
    return np.sqrt(n2)

def BK7(wl):
    #BK7
    n2 = 1 + (1.03961212*wl**2)/(wl**2-0.00600069867)+ \
         (0.231792344*wl**2)/(wl**2 - 0.0200179144) + \
         (1.01046945*wl**2)/(wl**2-103.560653)
    return np.sqrt(n2)
```

```python
wl = np.linspace(0.4,2,1000)
n_rhomb_val = BAF10(wl)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(wl,n_rhomb_val)
ax.set_title('Índice de refracción en función de la longitud de onda')
ax.set_xlabel('Longitud de onda ($\mu\:m$)',usetex=True)
ax.set_ylabel('Índice de refracción');
```

```python
n1 = 1.0
n2 = n_rhomb_val

# Entrada en el rombo
theta_i = np.radians(0.0)
theta_t = np.arcsin(n1/n2*np.sin(theta_i))
ts0 = 2*n1*np.cos(theta_i)/(n1*np.cos(theta_i)+n2*np.cos(theta_t))
tp0 = 2*n1*np.cos(theta_i)/(n2*np.cos(theta_i)+n1*np.cos(theta_t))
phase_s0 = np.angle(ts0)
phase_p0 = np.angle(tp0)

# Primera reflexion en el interior del rombo
theta_i = np.radians(60.0)
rs1 = (n1*np.cos(theta_i)-1j*n2*np.sqrt((n2/n1*np.sin(theta_i))**2-1))/ \
      (n1*np.cos(theta_i)+1j*n2*np.sqrt((n2/n1*np.sin(theta_i))**2-1))
rp1 = (n2*np.cos(theta_i)-1j*n1*np.sqrt((n2/n1*np.sin(theta_i))**2-1))/ \
      (n2*np.cos(theta_i)+1j*n1*np.sqrt((n2/n1*np.sin(theta_i))**2-1))
phase_s1 = np.angle(rs1)
phase_p1 = np.angle(rp1)

# Segunda reflexion en el interior del rombo
theta_i = np.radians(60.0)
rs2 = (n1*np.cos(theta_i)-1j*n2*np.sqrt((n2/n1*np.sin(theta_i))**2-1))/ \
      (n1*np.cos(theta_i)+1j*n2*np.sqrt((n2/n1*np.sin(theta_i))**2-1))
rp2 = (n2*np.cos(theta_i)-1j*n1*np.sqrt((n2/n1*np.sin(theta_i))**2-1))/ \
      (n2*np.cos(theta_i)+1j*n1*np.sqrt((n2/n1*np.sin(theta_i))**2-1))
phase_s2 = np.angle(rs2)
phase_p2 = np.angle(rp2)

# Salida en el rombo
theta_i = np.radians(0.0)
theta_t = np.arcsin(n2/n1*np.sin(theta_i))
ts3 = 2*n2*np.cos(theta_i)/(n2*np.cos(theta_i)+n1*np.cos(theta_t))
tp3 = 2*n2*np.cos(theta_i)/(n1*np.cos(theta_i)+n2*np.cos(theta_t))
phase_s3 = np.angle(ts3)
phase_p3 = np.angle(tp3)

phase_s = phase_s0 + phase_s1 + phase_s2 +phase_s3
phase_p = phase_p0 + phase_p1 + phase_p2 +phase_p3
phase = phase_s-phase_p
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.axhline(90,linestyle='--',c='k')
ax.plot(wl,np.degrees(phase))
# ax.set_ylim(0.4,0.6);
ax.set_ylim(88,92)
ax.set_xlim(0.3,2.1)
ax.set_title('Desfase a la salida en función de la longitud de onda')
ax.set_xlabel('Longitud de onda ($\mu\:m$)', usetex=True)
ax.set_ylabel('Desfase acumulado ($1/\pi$)')
```
